### Objective 1:
### Check the correlation between different types of animal farming practiced to make assumptions on its relation to number of households practicing agriculture.

```python
# Correlation Matrix
#use of a heat map to show correlation in the dataset.
# Correlation Matrix of the columns given below
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
cols =[' TOTAL HOUSHOLDS','Exotic cattle 0Dairy','Exotic cattle 0Beef','Exotic cattle 0Beef','Indigenous cattle','Sheep','Goats','Camels',
       'Donkeys','Pigs','Rabbits','Population (2019)','Area sq km','density per sq km']
f, ax = plt.subplots(figsize=(10, 7))
cm = np.corrcoef(Agric[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 10},
                 yticklabels=cols,
                 xticklabels=cols)
plt.show()
```

### Assumptions concluded
- Climate affects animal husbandry, this can be witnesses by the correlation between the area size and camels rearing. The increase in land size  bridges the gap of food shortage.
- Human population affects animal rearing, this can be seen by correlation between human population and the households practicing animal rearing to be strong and positive.
- Type of animal husbandry the animal rearing, eg beef farming where there is strong correlation between rabitts and cattle beef.
- Geographical area size affects animal husbandry. This can be drawn from correlation between area size and the  total households practicing the agriculture or correlation  between population density and total households.
